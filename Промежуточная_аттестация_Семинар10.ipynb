В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего
из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без
get_dummies?


```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
```

```python

# data = pd.DataFrame({'whoAmI'lst})
# data.head()
```

```python
data=pd.DataFrame(lst, columns=['whoAmI'])
```

```python
data
```

```python
lst_znach=list(set(lst))
```

```python
#print(lst_znach)
```

```python
lst_onehot=[]
for i in lst:
  lst_stroka=[]
  for j in lst_znach:
    if i==j:
      lst_stroka.append('1')
    else:
      lst_stroka.append('0')
  lst_onehot.append(lst_stroka)
```

```python
#print(lst)
#print(lst_onehot)
```

```python

data_new=pd.DataFrame(lst_onehot, columns=lst_znach)
data_new
```
